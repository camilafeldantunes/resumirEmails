```python
import os
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get('Gemini_APIkey')
```

```python
from google import genai

client = genai.Client()
```

```python
def resume_emails(emails):
  lista_de_resumos = []
  for i, elemento in enumerate(emails):
    chat = client.chats.create(model="gemini-2.5-flash")
    resposta = chat.send_message(elemento + ". Resuma esse email em apenas uma linha, definindo o que a pessoa queria no email")
    print(f"Email {i+1}: {resposta.text}")
    lista_de_resumos.append(f"Email{i+1}: {resposta.text}")
  return lista_de_resumos
```

```python
lista_resumos = resume_emails(emails)
```

```python
print(lista_resumos)
```

```python
with open("lista-de-resumos.txt", "w", encoding="utf-8") as arquivo:
  for elemento in lista_resumos:
    arquivo.write(elemento + "\n")

```
